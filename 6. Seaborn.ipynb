Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Importing the dataset
tip = sns.load_dataset('tips')
tip.head()
```

### Distribution Plot

```python
# Visualizing the distribution of data
sns.distplot(tip['total_bill'])
```

```python
sns.distplot(tip['total_bill'], kde = False)
# kde -> kernel density estimation 
# kde = False is used to remove it
```

```python
sns.distplot(tip['total_bill'], kde = False, bins =20)
```

### Joint Plot

```python
# In Join plot, we can join/merge two column & then compare them
sns.jointplot(x='total_bill', y='tip', data=tip)
# This give distribution plot and scatter plot as well
```

```python
sns.jointplot(x='total_bill', y='tip', data=tip, kind='hex')
# kind= hex is used for hexagonal shape
```

```python
sns.jointplot(x='total_bill', y='tip', data=tip, kind='reg')
# kind = reg is used for regression line
```

```python
sns.jointplot(x='total_bill', y='tip', data=tip, kind ='kde')
# kind =kde is used for density estimation -> gives the highest point where it matches up
```

### Pair plot

```python
Pair plot basically draws a relationship among different data frame pair-wise
```

```python
sns.pairplot(tip) # We get 9 different types of plot
```

```python
sns.pairplot(tip, hue ='sex') 
# Adding categorical variable(variable/column name having category, it should not be numerical)
# sex is only categorical variable -> do not contain numeric value, others are numerical 
```

```python
sns.pairplot(tip, hue='sex', palette='coolwarm') # palette is used to chnage the color
```

### Rug Plot

Rug plot is somehow similar to dist plot, where you can pass single column

```python
sns.rugplot(tip['total_bill'])
# Draws a dash mark at every single point for which that is along the dist line or distribution
```

```python
sns.rugplot(tip['tip'])
```
