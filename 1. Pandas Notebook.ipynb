# Slicing Subsets of Rows and Columns 

We can select specific ranges of our data in both the row and column directions using either label or integer-based indexing.

* loc is primarily label based indexing. Integers may be used but they are interpreted as a label.
* iloc is primarily integer based indexing

To select a subset of rows and columns from our DataFrame, we can use the iloc method. 
For example, we can select B and C (columns 1 and 2 if we start counting at 1)

```python
df
```

```python
# iloc[row slicing, column slicing]
df.iloc[0:2, 0:3]  # Slicing row 1 to row 2 & column 1 to cloumn 3 (column 1, column 2 & column 3)
```

```python

df.iloc[1:4, 1:3] # Counting starts from 1, so here row 2 & column 2
```

Indexing by labels loc differs from indexing by integers iloc. With loc, the both start bound and the stop bound are inclusive. When using loc, integers can be used, but the integers refer to the index label and not the position.

For example, using loc and select 1:2 will get a different result than using iloc to select rows 1:2

```python
# Selecting by row  label
df.loc['2' : '3']
```

```python
# Selecting by column label
df.loc[:,['A','C']] 
```

```python
# Specific row & column are selected by specifying row & column
df.loc[['1', '2','3'],['B','C']]
```

# Creating DataFrame

```python
# import pandas library to work with data
# create dataframe that we are importing

import pandas as pd

raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', 'Jacobson', ".", 'Milner', 'Cooze'], 
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, ".", "."],
        'postTestScore': ["25,000", "94,000", 57, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
# Save dataframe as csv file in the working directory
df.to_csv('example.csv')
```

```python
# Load/Read csv file
df = pd.read_csv('example.csv')
df
```

```python
print(df['age'].dtype)
```

```python
# load csv with no header
df = pd.read_csv('example.csv', header=None)
df
```

```python
# load csv specifying the column name
df = pd.read_csv('example.csv', names=['UID', 'First Name', 'Last Name', 'Age', 'Pre-Test Score', 'Post-Test Score'])
df
```

```python
# Load csv with setting the index column to UID
df = pd.read_csv('example.csv', index_col='UID', names=['UID', 'First Name', 'Last Name', 'Age', 'Pre-Test Score', 'Post-Test Score'])
df
```

```python
# Read csv file from row 3
df = pd.read_csv('example.csv', skiprows=3)
df
```
